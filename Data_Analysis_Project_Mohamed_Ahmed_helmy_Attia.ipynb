##  Define the missing values and update it with the mean for numerical values and the highest frequency for categorical values or delete it.

```python
import json
import csv
def read_people_data(data_in):
    x=0
    words_count=dict()
    people=dict()
    data=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','salary']
    for line in data_in:
            if len(line)>1:
                infoo=dict()
                i=0
                for word in line:
                    i=i%15
                    word=word.strip()
                    if word.isdigit():word=int(word)
                    if infoo.get(data[i],0)==0:infoo[data[i]]=[word]
                    else:infoo[data[i]].append(word)
                    if words_count.get(data[i],0)==0:words_count[data[i]]=[word]
                    else: words_count[data[i]].append(word)
                    i+=1
                people[f'{x}']=infoo
                x+=1
    return people,words_count


def cal_mean_freq(words_count):
    mean_freq=dict()
    sub_mean=dict()
    
    for keys,lists in words_count.items():
        total=0
        highest_freq_word=''
        sub_mean=dict()    
        for word in lists:        
            if isinstance(word,int):
                total+=word
                num=1
            else:
                num=0
                sub_mean[word]=sub_mean.get(word,0)+1        
        if num==0:            
            highest_val=0
            for k1,v1 in sub_mean.items():
                
                if v1>highest_val:
                    highest_val=v1                  
                    highest_freq_word=k1
            
            mean_freq[keys]=highest_freq_word
        else:mean_freq[keys]=(round(total/len(lists)))
        
    return mean_freq

def fix_missing(people1,mean_freq):   
    missing_data=list()
    for person,data in people1.items():
        for attribute,val in data.items():
            
            if val==['?']:
                replacement=mean_freq[attribute]
                text=f'Missing data from line {person}, Attribute {attribute}, Value ={val}, replaced with {replacement}'
                missing_data.append(text)
                people1[person][attribute]=replacement
                
    return people1,missing_data

def split_salary(people1):
    high_salary=dict()
    low_salary=dict()
    for person,data in people1.items():
        if data['salary']==['<=50K']:low_salary[person]=data
        else:high_salary[person]=data
    return high_salary,low_salary
def split_population(people1,word,value):
    matched=dict()
    nonmatched=dict()
    for person,data in people1.items():
        if data[word]==[value]:matched[person]=data
        else:nonmatched[person]=data
    return matched,nonmatched
def words_counter(people):
    counter=dict()
    
    data=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','salary']
    for word in data:
        sub_list=dict()
        counter[word]=[]
        num_list=list()
        for person,data in people.items():
            val=data[word]
            counter[word].append(val)
        for val in counter[word]:
            if word in ['age','fnlwgt','education-num','occupation','capital-gain','capital-loss','hours-per-week']:
                num=1
                
                num_list.append(val[0])
            else:
                num=0
                x=str(val[0])
                sub_list[x]=sub_list.get(x,0)+1
        if num==0:counter[word]=sub_list
        else: counter[word]=num_list
    return counter
def highest_lowest(words_counter,word):
    if word in ['age','fnlwgt','education-num','occupation','capital-gain','capital-loss','hours-per-week']:         
        highest_val=0
        lowest_val=9999999
        for val in words_counter[word]:
            if val>highest_val: highest_val=val
            if val<lowest_val:lowest_val=val
        return lowest_val,highest_val
    else:
        highest_val=0
        lowest_val=99999999
        for k1,v1 in words_counter[word].items():
            
            if v1>highest_val:
                highest_val=v1                  
                highest_freq_word=k1
            if v1<lowest_val:
                lowest_val=v1
                lowest_freq_word=k1
        return lowest_freq_word,highest_freq_word
```

```python
with open("adult.data", "r") as myfile:
        data_in = csv.reader(myfile, delimiter=',')
        people,words_count=read_people_data(data_in)
        mean_freq=cal_mean_freq(words_count)       
        people,missing_data=fix_missing(people,mean_freq)
        high_salary,low_salary=split_salary(people)
```

```python
# For The highest salary Group:
#a,b,c
high_salary_word_count=words_counter(high_salary)
high_salary_mean=cal_mean_freq(high_salary_word_count)

lowest_age,highest_age=highest_lowest(high_salary_word_count,'age')

lowest_education,highest_education=highest_lowest(high_salary_word_count,'education')

lowest_hours,highest_hours=highest_lowest(high_salary_word_count,'hours-per-week')

print(f'''Highest age is {highest_age}.\nLowest age is {lowest_age}.\nMean age is {high_salary_mean['age']}''')  
print('*'*20)
print(f'''Highest education is {highest_education}.\nLowest education is {lowest_education}.\nMean education is {high_salary_mean['education']}''') 
print('*'*20)
print(f'''Highest hours is {highest_hours}.\nLowest hours is {lowest_hours}.\nMean hours is {high_salary_mean['hours-per-week']}''')
```

```python
# For The lower salary Group:
#a,b,c
low_salary_word_count=words_counter(low_salary)
low_salary_mean=cal_mean_freq(low_salary_word_count)

lowest_age,highest_age=highest_lowest(low_salary_word_count,'age')

lowest_education,highest_education=highest_lowest(low_salary_word_count,'education')

lowest_hours,highest_hours=highest_lowest(low_salary_word_count,'hours-per-week')

print(f'''highest age is {highest_age}.\nLowest age is {lowest_age}.\nMean age is {low_salary_mean['age']}''')  
print('*'*20)
print(f'''highest education is {highest_education}.\nLowest education is {lowest_education}''') 
print('*'*20)
print(f'''highest hours is {highest_hours}.\nLowest hours is {lowest_hours}.\nMean hours is {low_salary_mean['hours-per-week']}''')
```
